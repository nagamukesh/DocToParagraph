Simple Model Example

```python
import transformers
from transformers import pipeline

# Load the pipeline
summarization_pipeline1 = pipeline("summarization", model="t5-base")
```

```python
# Define the text to summarize
text = "Hello guys, a very warm welcome and good morning to you all. This is a very very big passage with multiple unnecessary words which need to be removed from the sentence to make it small and summarize it. I really want to get a very beautiful output on the execution of this summarizer but this summarizer is not even changing the input which i had inputted, that is why, I am very dissapointed"

# Summarize the text
summary = summarization_pipeline1(text, max_length=50, min_length=0, do_sample=True)

# Print the summary
print(summary)
```

Now, using tokenizer to improve the model

```python
from transformers import AutoModelForSeq2SeqLM, AutoTokenizer

model = AutoModelForSeq2SeqLM.from_pretrained("t5-small")
tokenizer = AutoTokenizer.from_pretrained("t5-small")

summarization_pipeline2 = pipeline("summarization", model=model, tokenizer=tokenizer)
```

```python
/# Define the text to summarize
text = "Hello guys, a very warm welcome and good morning to you all. This is a very very big passage with multiple unnecessary words which need to be removed from the sentence to make it small and summarize it. I really want to get a very beautiful output on the execution of this summarizer but this summarizer is not even changing the input which i had inputted, that is why, I am very dissapointed"

# Summarize the text
summary = summarization_pipeline2(text, max_length=60, min_length=0, do_sample=True)

# Print the summary
print(summary[0]['summary_text'])
```
